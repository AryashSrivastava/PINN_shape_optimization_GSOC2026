### (iii) solving a 1D boundary-value ODE with a user-chosen forcing term. to solve $\frac{d^2u}{dx^2} = f(x)$ where I've chosen $f(x) = -sin(\pi x)$ as the forcing term. This is a classic problem (similar to steady-state heat conduction with a sinusoidal heat source).


```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
# Part 3
# Forcing term function: f(x) = -sin(pi * x)
```

```python
class SimplePINN(nn.Module):
    def __init__(self):
        super(SimplePINN, self).__init__()
        # Small, deep network: 1 input (x) -> 3 layers -> 1 output (u)
        self.net = nn.Sequential(
            nn.Linear(1, 50), nn.Tanh(),
            nn.Linear(50, 50), nn.Tanh(),
            nn.Linear(50, 1)
        )

    def forward(self, x):
        return self.net(x)
```

```python
model = SimplePINN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
# taking Forcing term function: f(x) = -sin(pi * x)
def f_forcing(x):
    return -torch.sin(np.pi * x)
```

```python
# 3. Training
epochs = 3000
x_domain = torch.linspace(0, 1, 100).view(-1, 1).requires_grad_(True)

for epoch in range(epochs):
    optimizer.zero_grad()
    #loss
    u = model(x_domain)
    # 1st derivative
    u_x = torch.autograd.grad(u, x_domain, torch.ones_like(u), create_graph=True)[0]
    #2nd derivative (Curvature)
    u_xx = torch.autograd.grad(u_x, x_domain, torch.ones_like(u_x), create_graph=True)[0]
    # ode: u_xx - f(x) = 0
    loss_physics = torch.mean((u_xx - f_forcing(x_domain))**2)
    #BOUndary loss (u(0)=0 and u(1)=0)
    u_boundary = model(torch.tensor([[0.0], [1.0]]))
    loss_boundary = torch.mean(u_boundary**2)
    #total Loss
    total_loss = loss_physics + loss_boundary
    total_loss.backward()
    optimizer.step()
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss = {total_loss.item():.6f}")
```

### (i) solving the Poisson equation on a 2D domain with user-specified boundary conditions; For this example, we'll use the forcing term $f(x, y) = -2\pi^2 \sin(\pi x) \sin(\pi y)$, which has the well-known analytical solution $u(x, y) = \sin(\pi x) \sin(\pi y)$ with zero Dirichlet boundary conditions on all sides.

```python
class PINN2D(nn.Module):
    def __init__(self):
        super(PINN2D, self).__init__()
        # 2 inputs (x, y) -> 3 hidden layers -> 1 output (u)
        self.net = nn.Sequential(
            nn.Linear(2, 64), nn.Tanh(),
            nn.Linear(64, 64), nn.Tanh(),
            nn.Linear(64, 64), nn.Tanh(),
            nn.Linear(64, 1)
        )

    def forward(self, x, y):
        return self.net(torch.cat([x, y], dim=1))
```

```python
# 1. Forcing Term and Exact Solution
def f_forcing(x, y):
    return -2 * np.pi**2 * torch.sin(np.pi * x) * torch.sin(np.pi * y)
def exact_solution(x, y):
    return np.sin(np.pi * x) * np.sin(np.pi * y)
```

```python
# 2. Setup Training Data
num_interior = 1000
num_boundary = 200
# Interior points (Collocation)
x_in = torch.rand(num_interior, 1, requires_grad=True)
y_in = torch.rand(num_interior, 1, requires_grad=True)
# Boundary points (x=0, x=1, y=0, y=1)
x_b = torch.cat([torch.zeros(num_boundary//4, 1), torch.ones(num_boundary//4, 1),
                 torch.rand(num_boundary//2, 1)])
y_b = torch.cat([torch.rand(num_boundary//2, 1), torch.zeros(num_boundary//4, 1),
                 torch.ones(num_boundary//4, 1)])
```

```python
# 3. Training Loop
model = PINN2D()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
epochs = 5000
loss_history = []
for epoch in range(epochs):
    optimizer.zero_grad()

    # Physics Loss (PDE)
    u = model(x_in, y_in)
    u_x = torch.autograd.grad(u, x_in, torch.ones_like(u), create_graph=True)[0]
    u_xx = torch.autograd.grad(u_x, x_in, torch.ones_like(u_x), create_graph=True)[0]
    u_y = torch.autograd.grad(u, y_in, torch.ones_like(u), create_graph=True)[0]
    u_yy = torch.autograd.grad(u_y, y_in, torch.ones_like(u_y), create_graph=True)[0]

    loss_pde = torch.mean((u_xx + u_yy - f_forcing(x_in, y_in))**2)

    # Boundary Loss (u=0)
    u_b = model(x_b, y_b)
    loss_bc = torch.mean(u_b**2)

    loss = loss_pde + 20 * loss_bc  # Weighting BCs higher helps convergence
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.6f}")
```

### (ii) solving the steady-state incompressible Navier–Stokes equations for flow around a specified geometry;

Solving the Navier–Stokes equations using PINNs is a significant step up in complexity from the Poisson equation. We transition from a single scalar field to a system of coupled PDEs involving a velocity vector field $\mathbf{u} = (u, v)$ and a pressure scalar field $p$.For this example, we will simulate 2D flow around a cylinder (a classic benchmark). The steady-state incompressible Navier–Stokes equations are:Momentum: $(\mathbf{u} \cdot \nabla)\mathbf{u} + \frac{1}{\rho}\nabla p - \nu \Delta \mathbf{u} = 0$Continuity: $\nabla \cdot \mathbf{u} = 0$

```python
class NavStokesPINN(nn.Module):
    def __init__(self):
        super(NavStokesPINN, self).__init__()
        # Input: (x, y) -> Output: (u, v, p)
        self.net = nn.Sequential(
            nn.Linear(2, 64), nn.Tanh(),
            nn.Linear(64, 64), nn.Tanh(),
            nn.Linear(64, 64), nn.Tanh(),
            nn.Linear(64, 64), nn.Tanh(),
            nn.Linear(64, 3)
        )

    def forward(self, xy_input): # Changed to accept a single concatenated xy tensor
        res = self.net(xy_input)
        return res[:, 0:1], res[:, 1:2], res[:, 2:3] # u, v, p
```

```python
# 1. Constants (Low Reynolds Number for stability)
nu = 0.01  # Kinematic viscosity
rho = 1.0
```

```python
def pde_residuals(model, x, y):
    # Combine x and y into a single input tensor for the model
    xy_input = torch.cat([x, y], dim=1) # x and y already require_grad=True

    u, v, p = model(xy_input)

    # Compute first-order derivatives for u, v, p with respect to xy_input
    # grad_u_xy will have shape (N, 2), where column 0 is du/dx and column 1 is du/dy
    grad_u_xy = torch.autograd.grad(u, xy_input, grad_outputs=torch.ones_like(u), create_graph=True)[0]
    u_x, u_y = grad_u_xy[:, 0:1], grad_u_xy[:, 1:2]

    grad_v_xy = torch.autograd.grad(v, xy_input, grad_outputs=torch.ones_like(v), create_graph=True)[0]
    v_x, v_y = grad_v_xy[:, 0:1], grad_v_xy[:, 1:2]

    grad_p_xy = torch.autograd.grad(p, xy_input, grad_outputs=torch.ones_like(p), create_graph=True)[0]
    p_x, p_y = grad_p_xy[:, 0:1], grad_p_xy[:, 1:2]

    # Compute second-order derivatives
    # To get u_xx, we differentiate u_x with respect to xy_input and take the x-component of the result
    u_xx = torch.autograd.grad(u_x, xy_input, grad_outputs=torch.ones_like(u_x), create_graph=True)[0][:, 0:1]
    u_yy = torch.autograd.grad(u_y, xy_input, grad_outputs=torch.ones_like(u_y), create_graph=True)[0][:, 1:2]

    v_xx = torch.autograd.grad(v_x, xy_input, grad_outputs=torch.ones_like(v_x), create_graph=True)[0][:, 0:1]
    v_yy = torch.autograd.grad(v_y, xy_input, grad_outputs=torch.ones_like(v_y), create_graph=True)[0][:, 1:2]

    # Navier-Stokes Residuals
    f_u = u*u_x + v*u_y + (1/rho)*p_x - nu*(u_xx + u_yy)
    f_v = u*v_x + v*v_y + (1/rho)*p_y - nu*(v_xx + v_yy)
    f_cont = u_x + v_y # Continuity equation

    return f_u, f_v, f_cont
```

```python
# 2. Setup Training (Simplified Domain [0, 2]x[0, 1] with cylinder at (0.5, 0.5))
# In a real scenario, you'd mask out the interior of the cylinder.
x_coll = torch.rand(2000, 1, requires_grad=True) * 2.0
y_coll = torch.rand(2000, 1, requires_grad=True)

# Re-instantiate the model to pick up the latest NavStokesPINN class definition
model = NavStokesPINN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0008)
```

```python
# 3. Training (Simplified Logic)
for epoch in range(3001):
    optimizer.zero_grad()

    # PDE Loss
    # x_coll and y_coll are already created with requires_grad=True
    fu, fv, fc = pde_residuals(model, x_coll, y_coll)
    loss_pde = torch.mean(fu**2 + fv**2 + fc**2)

    # Boundary Loss (Inlet u=1, v=0 at x=0)
    inlet_x = torch.zeros(100, 1)
    inlet_y = torch.rand(100, 1)
    inlet_points = torch.cat([inlet_x, inlet_y], dim=1) # Combine for model input
    u_in, v_in, _ = model(inlet_points)
    loss_bc = torch.mean((u_in - 1.0)**2 + v_in**2)

    loss = loss_pde + 10 * loss_bc
    loss.backward(retain_graph=True)
    optimizer.step()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.6f}")
```

```python
def visualize_flow(model, device='cpu'):
    # Create a grid for testing
    nx, ny = 200, 100
    x = np.linspace(0, 2.0, nx)
    y = np.linspace(0, 1.0, ny)
    X, Y = np.meshgrid(x, y)

    X_flat = torch.tensor(X.flatten()[:, None], dtype=torch.float32)
    Y_flat = torch.tensor(Y.flatten()[:, None], dtype=torch.float32)
    xy_input = torch.cat([X_flat, Y_flat], dim=1)

    with torch.no_grad():
        u, v, p = model(xy_input)
        u = u.reshape(ny, nx).numpy()
        v = v.reshape(ny, nx).numpy()
        p = p.reshape(ny, nx).numpy()
        speed = np.sqrt(u**2 + v**2)

    # Masking the cylinder area for visualization (Center: 0.5, 0.5, Radius: 0.1)
    mask = (X - 0.5)**2 + (Y - 0.5)**2 < 0.1**2
    u[mask] = np.nan
    v[mask] = np.nan
    speed[mask] = np.nan

    fig, ax = plt.subplots(3, 1, figsize=(12, 12))

    # Plot Velocity Magnitude (Speed)
    im0 = ax[0].contourf(X, Y, speed, levels=50, cmap='jet')
    ax[0].set_title("Velocity Magnitude (Speed)")
    fig.colorbar(im0, ax=ax[0])

    # Plot Pressure Field
    im1 = ax[1].contourf(X, Y, p, levels=50, cmap='coolwarm')
    ax[1].set_title("Pressure Field")
    fig.colorbar(im1, ax=ax[1])

    # Plot Streamlines
    # Note: Streamlines don't like NaNs, so we use the unmasked data for calculation
    ax[2].streamplot(x, y, u, v, color='black', linewidth=1, density=1.5)
    ax[2].set_title("Flow Streamlines")
    # Draw the cylinder
    circle = plt.Circle((0.5, 0.5), 0.1, color='gray', zorder=5)
    ax[2].add_artist(circle)

    plt.tight_layout()
    plt.show()

# Call the function after training
visualize_flow(model)
```
